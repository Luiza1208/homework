```python
movielist = soup1.findAll('span',class_ = 'styles_mainTitle__IFQyZ styles_activeMovieTittle__kJdJj')
originl_name_years = soup1.findAll('span',class_ = 'desktop-list-main-info_secondaryText__M_aus')
rating = soup1.findAll('span',class_ = 'styles_kinopoiskValuePositive__vOb2E styles_kinopoiskValue__9qXjg')
country = soup1.findAll('span', class_ = 'desktop-list-main-info_truncatedText__IMQRP')
position = soup1.findAll('span',class_ = 'styles_position__TDe4E')
```
